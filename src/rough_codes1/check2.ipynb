```python
filename = input('Enter a filename: ')
with open('human/'+filename, 'r') as file:
 data = file.read()
```

```python
import spacy
nlp = spacy.load("en_core_web_lg")
```

```python
str1=" ".join(data.split())
```

WORD LENGTH

```python
text = nlp(str1)
sent_tokenize = (list(text.sents))
i=0
avg=0
mode=[]
for sent in sent_tokenize:
    i+=1
    print (len(sent))
    avg+=len(sent)
    mode.append(len(sent))
avg=avg/i
print(avg)
print(i)
```

```python
import statistics
print(statistics.mode(mode))
```

Sentence similarity

```python
text = nlp(str1)
sent_tokenize = (list(text.sents))
i=0
sensim=0.0
for sent2 in sent_tokenize:
    i+=1
    if i == 1:
        sent1=sent2
        continue
    print (sent2.similarity(sent1))
    sent1=sent2
    sensim+=sent2.similarity(sent1)
sensim=sensim/i
print("average sentence similarity is ",sensim)
```

word similarity

```python
#without stopwords
words = [token.text for token in text if token.is_stop != True and token.is_punct != True]

for i in range(len(words)):
    #print(i)
    #print(words[i])
    w2=words[i]
    if i == 0:
        w1=w2
        continue
    print (nlp(w2).similarity(nlp(w1)))
    print(w1+ " "+w2)
    w1=w2
```
