```python
import matplotlib.pyplot as plt
import numpy as np

for index in range(2,num_columns):
    fig = plt.figure(figsize=(14, 6))
    feature=df1.columns[index]
# Get the row index
    row_index = df2.index

# Get the values of the features
    values1 = df1.iloc[:,index]
    values2 = df2.iloc[:,index]

    # Generate random data for two datasets
    data1 = np.random.normal(values1.mean(), values1.std(), 150)
    data2 = np.random.normal(values2.mean(), values2.std(), 150)

# Plot the histograms
    plt.hist(data1, bins=30,color='skyblue', edgecolor='black', alpha=0.5, label='gpt')
    plt.hist(data2, bins=30,color='pink', edgecolor='black', alpha=0.5, label='human')
# Set the labels and title
    plt.xlabel('Values')
    plt.ylabel('counts')
    plt.title('Comparison of Features: '+feature)
    plt.legend()

# Show the plot
    plt.show()
    #ch=input("enter c to continue and e to exit")
    #if ch != 'c':
    #    break
```
