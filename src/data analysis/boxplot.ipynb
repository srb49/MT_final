```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches
```

```python
df1 = pd.read_csv('../../csv/gpt/csv_gpt3-2.csv')
df2 = pd.read_csv('../../csv/human/csv_human-2.csv')
```

```python
plt.scatter(df1.iloc[:,20], df2.iloc[:,20])
print(df1.iloc[0,20], df2.iloc[0,21])
plt.xlabel(df1.columns[20])
plt.ylabel(df2.columns[20])
plt.title('Comparison of Feature 21')
plt.show()
```

```python
l=df1.shape[1]
print(l)
```

```python
l=df1.shape[1]
for i in range(3,l-2): 
   # Scatter plot for dataset 1 (blue points)
    plt.scatter(df1.iloc[:,i], df1.iloc[:,36], c='skyblue', label='gpt')

    # Scatter plot for dataset 2 (red points)
    plt.scatter(df2.iloc[:,i], df2.iloc[:,36], c='pink', label='human')

    plt.xlabel(df1.columns[i])
    plt.ylabel(df2.columns[i+1])
    #plt.title('Comparison of Feature ',df1.columns[i], 'vs ',df1.columns[i+1] )
    plt.show()
```

```python
l=df1.shape[1]
for i in range(2,l-2): 
   # Scatter plot for dataset 1 (blue points)
    plt.scatter(df1.iloc[:,i], df1.iloc[:,i+1], c='skyblue', label='gpt')

    # Scatter plot for dataset 2 (red points)
    plt.scatter(df2.iloc[:,i], df2.iloc[:,i+1], c='pink', label='human')

    plt.xlabel(df1.columns[i])
    plt.ylabel(df2.columns[i+1])

    legend_elements = [mpatches.Patch(color='skyblue', label='gpt'),
                       mpatches.Patch(color='pink', label='human')]

# Add the legend with custom elements
    plt.legend(handles=legend_elements)
    #plt.title('Comparison of Feature ',df1.columns[i], 'vs ',df1.columns[i+1] )
    plt.show()
```
